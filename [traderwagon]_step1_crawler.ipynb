```python
import pymongo
import os
import json
import requests
from pprint import pprint as pp
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
MONGO_URL = os.environ.get('MONGO_URL')
DB_NAME = MONGO_URL.split('/')[-1]
client = pymongo.MongoClient(os.environ.get('MONGO_URL'))
db = client.get_database(DB_NAME)
```

```python
def get_list_lead_port_order_perf(port_id):
    url = "https://www.traderwagon.com/v1/public/social-trading/lead-portfolio/list-lead-portfolio-order-performance"

    payload = json.dumps({
      "page": 1,
      "rows": 1000,
      "leadPortfolioId": f"{port_id}"
    })
    headers = {
      'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:109.0) Gecko/20100101 Firefox/113.0',
      'Accept': '*/*',
      'Accept-Language': 'en-US,en;q=0.5',
      'Accept-Encoding': 'gzip, deflate, br',
      'lang': 'en',
      'content-type': 'application/json',
      'clienttype': 'web',
      'Origin': 'https://www.traderwagon.com',
      'Connection': 'keep-alive',
      'Sec-Fetch-Dest': 'empty',
      'Sec-Fetch-Mode': 'cors',
      'Sec-Fetch-Site': 'same-origin',
      'Pragma': 'no-cache',
      'Cache-Control': 'no-cache',
      'TE': 'trailers'
    }

    response = requests.request("POST", url, headers=headers, data=payload)
    return response.json()
```

```python
for i in range(9999, 1, -1):
  print(f"fetching order history from port {i}")
  result = get_list_lead_port_order_perf(i)
  if len(result['data']):
    insert_id = db.orders.insert_many(result['data'])
    print(f"finished insert for port {i}, {insert_id.inserted_ids}")
```
